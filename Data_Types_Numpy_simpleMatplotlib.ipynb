```python
import numpy as np
```

```python
some_array = np.array([[1,2,3,4,5],[2,3,6,5,4]])
some_array.dtype
some_array.ndim
some_array.itemsize
```

```python
some_array=np.ones((3,5))
some_array
```

```python
some_array=np.zeros((3,5))
some_array
```

```python
some_array=np.empty((3,5))
some_array
```

```python
some_array=np.arange(1,10,2)# start,end,space
some_array
```

```python
some_array=np.linspace(1,10,150)# start,end,no of data points including start and end
some_array
```

```python
some_array = np.random.normal(0,1,100)
some_array
```

```python
from matplotlib import pyplot as plt
```

```python
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(some_array,label='random series')
plt.legend()
plt.savefig('random.jpg')
```

```python
sorted_some_array=np.sort(some_array)
```

```python
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(sorted_some_array,label='sorted_random series')
plt.legend()
```

```python
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(sorted_some_array+ 0.26*some_array,label='noisey series')
plt.legend()
```

```python
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(np.sort(sorted_some_array+ 0.26*some_array),label='sorted noisey series')
plt.legend()
```

```python
plt.subplot(2,2,1)

plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(some_array,label='random series')
plt.legend()

plt.subplot(2,2,2)

plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(sorted_some_array,label='sorted_random series')
plt.legend()

plt.subplot(2,2,3)

plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(sorted_some_array+ 0.26*some_array,label='noisey series')
plt.legend()

plt.subplot(2,2,4)

plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.plot(np.sort(sorted_some_array+ 0.26*some_array),label='sorted noisey series')
plt.legend()
```
