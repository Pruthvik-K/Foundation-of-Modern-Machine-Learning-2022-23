```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
test_sample = np.linspace(1,10,2000)
```

```python
plt.plot(np.sin(test_sample))
plt.plot(np.cos(test_sample))
plt.savefig('someplots.jpg')
```

```python
first_fig = cv.imread('someplots.jpg',1)
```

```python
rows,cols,depth = first_fig.shape
```

```python
hstacked_fig = np.hstack((first_fig, first_fig))
plt.imshow(hstacked_fig)
plt.savefig('hstacked_image.jpg')
```

```python
vstacked_fig = np.vstack((hstacked_fig,hstacked_fig))
```

```python
vstacked_fig.shape
```

```python
plt.imshow(vstacked_fig)
plt.savefig('vstacked_image.jpg')
```

```python
portion = vstacked_fig[0:rows, 0:cols, 0:depth]
```

```python
plt.imshow(portion)
```

```python
blured_portion = cv.blur(portion, (10,10))
plt.savefig('blured_image.jpg')
```

```python
plt.imshow(blured_portion)
```

```python
vstacked_fig[0:rows, 0:cols, 0:depth] = blured_portion
```

```python
plt.imshow(vstacked_fig)
plt.savefig('stacked_blured_image.jpg')
```
